```python
#Removing outliers

#For 'P' column

Q1=np.quantile(df['P'],0.25)
Q3=np.quantile(df['P'],0.75)

IQR=Q3-Q1

UQR=Q3+1.5*IQR
LQR=Q1-1.5*IQR

df=df[(df['P']<=UQR) & (df['P']>=LQR)]

#For 'K' column

Q1=np.quantile(df['K'],0.25)
Q3=np.quantile(df['K'],0.75)

IQR=Q3-Q1

UQR=Q3+1.5*IQR
LQR=Q1-1.5*IQR

df=df[(df['K']<=UQR) & (df['K']>=LQR)]

#For 'temperature' column

Q1=np.quantile(df['temperature'],0.25)
Q3=np.quantile(df['temperature'],0.75)

IQR=Q3-Q1

UQR=Q3+1.5*IQR
LQR=Q1-1.5*IQR

df=df[(df['temperature']<=UQR) & (df['temperature']>=LQR)]


#For 'ph' column

Q1=np.quantile(df['ph'],0.25)
Q3=np.quantile(df['ph'],0.75)

IQR=Q3-Q1

UQR=Q3+1.5*IQR
LQR=Q1-1.5*IQR

df=df[(df['ph']<=UQR) & (df['ph']>=LQR)]


#For 'rainfall' column

Q1=np.quantile(df['rainfall'],0.25)
Q3=np.quantile(df['rainfall'],0.75)

IQR=Q3-Q1

UQR=Q3+1.5*IQR
LQR=Q1-1.5*IQR

df=df[(df['rainfall']<=UQR) & (df['rainfall']>=LQR)]
```

```python
#Converting categorical label to numberic values

from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()
df['label']=lb.fit_transform(df['label'])
```

```python
#Taking independant and independant variables

x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#Splitting data into train and test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('Shape of train:',x_train.shape, y_train.shape)
print('Shape of test:',x_test.shape, y_test.shape)
```

```python
#Standardizing data

from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
#Model building

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
models =[("SVC", SVC()),
         ("DecisionTree", DecisionTreeClassifier()),
         ("NaiveBayes", GaussianNB()),
         ('RandomForest',RandomForestClassifier()),
         ('GradientBoost',GradientBoostingClassifier()),
         ('XGBClassifier',XGBClassifier()),
         ('KNeighborsClassifier',KNeighborsClassifier(n_neighbors=5))]
```

```python
#Calculating accuracy of different models
names = []
accuracies = []
finalResults = []
for name,model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracyScore = accuracy_score(y_test, y_pred)
    accuracies.append(accuracyScore)
    names.append(name)
    finalResults.append((name,accuracyScore))
```

```python
finalResults_df = pd.DataFrame({'Model-Name': names, 'Accuracy-Score': accuracies})
finalResults_df.sort_values(by='Accuracy-Score', ascending=False)
```
